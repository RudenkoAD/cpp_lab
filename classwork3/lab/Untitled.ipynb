```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('experiment5.csv', sep=';')
data = data.transpose()
data
```

```python
vavg=[]
x=[]
for i in range(2020):
  vavg.append(np.average(data[i][2:1000]))
  x.append(i%101)
x=x[1:]
vavg=vavg[1:]
```

```python
plt.plot(x, vavg, 'o')

x_range=np.arange(0, 100, 0.01)
y_range=[np.polyval(a, x) for x in x_range]
plt.plot(x_range, y_range, color='red')
```

```python
for size in range(1, 21):
    plt.plot(x[(size-1)*101:(size)*101-1], vavg[(size-1)*101:(size)*101-1])
    plt.show()
```
